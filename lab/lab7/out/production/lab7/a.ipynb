```python
import matplotlib.pyplot as plt

# 假设你有一些数据，这里我使用简单的指数函数来模拟运行时间与问题规模的关系
# 你需要将这里的数据替换为你实际的数据
def runtime_function(n):
    return 0.0005*2 ** n

# 生成一些问题规模 n 的值
n_values = list(range(1, 21))  # 假设问题规模从 1 到 20

# 计算每个问题规模对应的运行时间
runtimes = [runtime_function(n) for n in n_values]

# 绘制图表
plt.plot(n_values, runtimes, marker='o', linestyle='-')
plt.xlabel('n')
plt.ylabel('time')
plt.title('Runtime of a function of n')
plt.grid(True)
plt.show()
```
