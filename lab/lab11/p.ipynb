```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 14})
# 定义要最小化的函数
def f(x):
    return ((x[0]+1)**2)/9 + (x[1]+1)**2

# 定义函数的梯度
def gradient_f(x):
    return np.array([(2/9)*(x[0]+1), 2*(x[1]+1)])

# 带动量的梯度下降
def gradient_descent_with_momentum(initial_x, learning_rate, momentum, num_iterations):
    x = initial_x
    velocity = np.zeros_like(x)
    x_history = [x]
    
    for _ in range(num_iterations):
        grad = gradient_f(x)
        velocity = momentum * velocity + learning_rate * grad
        x = x - velocity
        x_history.append(x)
        
    return x_history

# 不带动量的梯度下降
def gradient_descent(initial_x, learning_rate, num_iterations):
    x = initial_x
    x_history = [x]
    
    for _ in range(num_iterations):
        grad = gradient_f(x)
        x = x - learning_rate * grad
        x_history.append(x)
        
    return x_history

# 设置初始参数
initial_x1 = np.array([-3.0, -3.0])
initial_x2 = np.array([1.0, 1.0])
learning_rate = 0.8
momentum = 0.6  # 恒定的动量项
num_iterations = 20

# 执行带动量的梯度下降
trajectory_with_momentum = gradient_descent_with_momentum(initial_x1, learning_rate, momentum, num_iterations)

# 绘制带动量的梯度下降的优化轨迹
trajectory_with_momentum = np.array(trajectory_with_momentum)
x_values_with_momentum = trajectory_with_momentum[:, 0]
y_values_with_momentum = trajectory_with_momentum[:, 1]

# 执行不带动量的梯度下降
trajectory_no_momentum = gradient_descent(initial_x2, learning_rate, num_iterations)

# 绘制不带动量的梯度下降的优化轨迹
trajectory_no_momentum = np.array(trajectory_no_momentum)
x_values_no_momentum = trajectory_no_momentum[:, 0]
y_values_no_momentum = trajectory_no_momentum[:, 1]

# 绘制函数的等高线图
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = ((X+1)**2)/9 + (Y+1)**2
plt.contour(X, Y, Z, levels=20)

# 绘制带动量的梯度下降的轨迹
plt.plot(x_values_with_momentum, y_values_with_momentum, '-o', color='r', label='with momentum')
# 绘制不带动量的梯度下降的轨迹
plt.plot(x_values_no_momentum, y_values_no_momentum, '-o', color='b', label='without momentum')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Gradient Descent with Momentum')
plt.legend()

# plt.grid(True)
plt.show()
plt.close()
```
